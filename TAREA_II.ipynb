```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import urllib.parse

# ESPERAR A QUE LA TABLA DENTRO DE 'DIVDETALLE' SEA VISIBLE

wait.until(EC.visibility_of_element_located((By.XPATH, "//div[@id='divDetalle']/table/tbody/tr/td/table")))

# ENCONTRAR LA TABLA

table_element = driver.find_element(By.XPATH, "//div[@id='divDetalle']/table/tbody/tr/td/table")

# DEFINIMOS LOS ENCABEZADOS MANUALMENTE PARA QUE COINCIDAN CON LAS 11 COLUMNAS DE DATOS

headers = ['Nro', 'Año', 'Período', 'Código Encuesta', 'Encuesta', 'Código Módulo', 'Módulo', 'Ficha', 'Descarga SPSS', 'Descarga STATA', 'Descarga CSV']

# EXTRAE LAS FILAS DE DATOS A PARTOR DE LA SEGUNDA FILA DEL TBODY DE LA TABLA INTERNA

data = []
base_url = "https://proyectos.inei.gob.pe"

for row in table_element.find_elements(By.XPATH, ".//tbody/tr[position() > 1]"):
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = []
    for i, cell in enumerate(cells):
        if i in [8, 9, 10]:  # COLUMNAS (SPSS, STATA, CSV)
            try:
                link_element = cell.find_element(By.TAG_NAME, 'a')
                href_value = link_element.get_attribute('href')


                full_url = urllib.parse.urljoin(base_url, href_value)


                if full_url.startswith(base_url + base_url):
                    full_url = base_url + full_url[len(base_url):]

                row_data.append(full_url)
            except:
                row_data.append('')
        else:
            row_data.append(cell.text)
    data.append(row_data)

# CREACIÓN DE UN DATAFRMA DE PANDAS

df = pd.DataFrame(data, columns=headers)

# MOSTRAMOS LAS PRIMERAS FILAS DEL DATAFRAME

print("Datos extraídos de la tabla:")
display(df.head())

# DEFINIR LA RUTA DE GUARDADO DE LA TABLA

drive_path = '/content/drive/MyDrive/Q_LAB/Especialización/Tarea_II/extracted_table.csv'

# EXTRA: CREAR DIRECTORIO SI NO EXISTE

os.makedirs(os.path.dirname(drive_path), exist_ok=True)

# GUARDAR EL DATAFRAME COMO CSV EN GOOGLE DRIVE

df.to_csv(drive_path, index=False)
print(f"Tabla guardada exitosamente en: {drive_path}")
```

```python
import requests
import os

# MÓDULOS QUE EL USUARIO DESEA DESCARGAR

modules_to_download = [
    'Salud',
    'Educación',
    'Empleo e Ingresos',
    'Características de la Vivienda y del Hogar',
    'Características de los Miembros del Hogar'
]

# RUTA DE GUARDADO EN GOOGLE DRIVE

target_drive_path = '/content/drive/MyDrive/Q_LAB/Especialización/Tarea_II/ENAHO_2024'

# CREAMOS EL DIRECTORIO SI ES QUE NO EXISTE O NO RECONOCE

os.makedirs(target_drive_path, exist_ok=True)
print(f"Directorio de destino creado: {target_drive_path}")

# FILTRAR EL DATAFRAME PARA OBTENER SOLO LOS MÓDULOS DESEADOS CON ENLACES CSV

filtered_df = df[df['Módulo'].isin(modules_to_download) & (df['Descarga CSV'] != '')]

if filtered_df.empty:
    print("No se encontraron enlaces CSV para los módulos solicitados.")
else:
    print(f"Se encontraron {len(filtered_df)} archivos CSV para descargar.")

    # ITERAR SOBRE LAS FILAS FILTRADAS Y DESCARGA CADA ARCHIVO

    for index, row in filtered_df.iterrows():
        module_name = row['Módulo']
        csv_url = row['Descarga CSV']

        # CREACIÓN DE UN NOMBRE DE ARCHIVO SEGURO Y DESCRIPTIVO
        # REEMPLAZO DE CARACTERES ESPECIALES Y ESPACIOS CON GUIONES BAJOS

        safe_module_name = module_name.replace(' ', '_').replace('/', '_').replace(':', '_').replace('-', '_').replace('(', '').replace(')', '')
        file_name = f"ENAHO_2024_{safe_module_name}.zip"
        file_path = os.path.join(target_drive_path, file_name)

        print(f"Intentando descargar '{module_name}' desde: {csv_url}")
        try:
            response = requests.get(csv_url, stream=True)
            response.raise_for_status()

            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=8192):
                    f.write(chunk)
            print(f"Descargado y guardado: {file_path}")
        except requests.exceptions.RequestException as e:
            print(f"Error al descargar '{module_name}': {e}")
        except Exception as e:
            print(f"Ocurrió un error inesperado al procesar '{module_name}': {e}")

print("Proceso de descarga completado.")
```

### 5. Extracción de archivos CSV de los ZIPs descargados

Ahora que tenemos los archivos ZIP, procederemos a extraer los archivos CSV específicos que necesitas para tu análisis. Crearemos una nueva carpeta para los CSV extraídos.

```python
import zipfile
import os

# RUTA DONDE SE DESCARGARON LOS ARCHIVOS ZIP

source_zip_path = '/content/drive/MyDrive/Q_LAB/Especialización/Tarea_II/ENAHO_2024'

# RUTA DONDE SE GUARDARÁN LOS ARCHIVOS CSV EXTRAÍDOS

target_csv_path = '/content/drive/MyDrive/Q_LAB/Especialización/Tarea_II/ENAHO_2024_CSVs_Extraidos'

# LISTA DE NOMBRES DE ARCHIVOS CSV A EXTRAER

csv_files_to_extract = [
    'Enaho01-2024-100.csv',
    'Enaho01-2024-200.csv',
    'Enaho01A-2024-300.csv',
    'Enaho01a-2024-500.csv',
    'Enaho01A-2024-400.csv'
]

# CREAR EL DIRECTORIO DE DESTINO PARA LOS CSV SI NO EXISTE

os.makedirs(target_csv_path, exist_ok=True)
print(f"Directorio para CSVs extraídos creado: {target_csv_path}")

# OBTENER UNA LISTA DE TODOS LOS ARCHIVOS ZIP EN LA CARPETA DE ORIGEN

zip_files = [f for f in os.listdir(source_zip_path) if f.endswith('.zip')]

if not zip_files:
    print(f"No se encontraron archivos ZIP en: {source_zip_path}")
else:
    print(f"Se encontraron {len(zip_files)} archivos ZIP. Procesando...")
    for zip_file_name in zip_files:
        zip_file_full_path = os.path.join(source_zip_path, zip_file_name)
        print(f"\nProcesando archivo ZIP: {zip_file_name}")
        try:
            with zipfile.ZipFile(zip_file_full_path, 'r') as zip_ref:

                # LISTA DE ARCHIVOS DENTRO DEL ZIP

                list_of_files_in_zip = zip_ref.namelist()

                extracted_any = False
                for csv_target_name in csv_files_to_extract:

                    # BUSCAR ARCHIVOS CON COINCIDENCIA

                    found_csv_in_zip = None
                    for file_in_zip in list_of_files_in_zip:

                        # NORMALIZAMOS LOS NOMBRES

                        normalized_file_in_zip = file_in_zip.lower()
                        normalized_csv_target_name = csv_target_name.lower()

                        if normalized_file_in_zip.endswith(normalized_csv_target_name):
                            found_csv_in_zip = file_in_zip
                            break

                    if found_csv_in_zip:
                        print(f"  Extrayendo '{found_csv_in_zip}'...")
                        zip_ref.extract(found_csv_in_zip, target_csv_path)

                        # RENOMBRAR SI ES NECESARIO ASEGURAR EL NOMBRE DESEADO

                        extracted_file_path = os.path.join(target_csv_path, found_csv_in_zip)
                        final_csv_name = os.path.join(target_csv_path, csv_target_name)
                        if extracted_file_path != final_csv_name:
                            os.rename(extracted_file_path, final_csv_name)
                        print(f"    Guardado como: {final_csv_name}")
                        extracted_any = True
                    else:
                        print(f"  Advertencia: No se encontró '{csv_target_name}' (o su variante) en el ZIP '{zip_file_name}'.")

                if not extracted_any:
                    print(f"  No se extrajo ningún archivo CSV solicitado del ZIP '{zip_file_name}'.")

        except zipfile.BadZipFile:
            print(f"  Error: '{zip_file_name}' no es un archivo ZIP válido o está corrupto.")
        except Exception as e:
            print(f"  Ocurrió un error inesperado al procesar '{zip_file_name}': {e}")

print("\nProceso de extracción completado.")
```
